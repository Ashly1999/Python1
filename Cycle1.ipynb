```
#Program to Calculate the Sum of Natural Numbers
num = int(input("Enter a number: "))  
  
if num < 0:  
   print("Enter a positive number")  
else:  
   sum = 0  
   # use while loop to iterate un till zero  
   while(num > 0):  
       sum += num  
       num -= 1  
   print("The sum is",sum)    
   
```

```
#Program to find Factorial of a Number
num = int(input("Enter a number: "))  
factorial = 1  
if num < 0:  
   print("Sorry, factorial does not exist for negative numbers")  
elif num == 0:  
   print("The factorial of 0 is 1")  
else:  
   for i in range(1,num + 1):  
       factorial = factorial*i  
   print("The factorial of",num,"is",factorial)  
```

```
#Program to display Fibonacci Sequence
n=int(input("Enter any number:"))
x,y=0,1
count=0
if n<=0:
   print("Enter a positive number:")
elif n==1:
      print("Fibonacci sequence upto",n,":")
      print(x)
else:
      print("Fibonacci sequence:")
      while count<n:
            print(x)
            z=x+y
            x=y
            y=z
            count+=1      
        
```

```
#Program to find LCM of two numbers
a=int(input("Enter the first number:"))
b=int(input("Enter the second number:"))
if(a>b):
    min1=a
else:
    min1=b
while(1):
    if(min1%a==0 and min1%b==0):
        print("LCM is:",min1)
        break
    min1=min1+1
```

```
#Program to reverse of a number
Number = int(input("Please Enter any Number: "))
Reverse = 0
while(Number > 0):
    Reminder = Number %10
    Reverse = (Reverse *10) + Reminder
    Number = Number //10

print(" Reverse of entered number is ",Reverse)
```

```
#Program to calculate the power of a number
number = int(input("  Enter any Positive Integer : "))
exponent = int(input("  Enter Exponent Value : "))
power = 1

for i in range(1, exponent + 1):
    power = power * number
    
print("The Result of {0} Power {1} = {2}".format(number, exponent, power))
```

```
#Program to check whether a number is Palindrome
number = int(input("Please Enter any Number: "))

reverse = 0
temp = number

while(temp > 0):
    Reminder = temp % 10
    reverse = (reverse * 10) + Reminder
    temp = temp //10
 
print("Reverse of a Given number is " ,reverse)

if(number == reverse):
    print("is a Palindrome Number" ,number)
else:
    print(" is not a Palindrome Number" ,number)
             
```
